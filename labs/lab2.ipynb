Каков эквивалент функции enumerate для numpy массивов?


```python
arr22 = np.random.randint(0,16, 16)
for key, value in enumerate(arr22):
    print(key, value)
print()
#вариант 1
for key, value in np.ndenumerate(arr22):
    print(key, value)
print()
#вариант 2
for key in np.ndindex(arr22.shape):
    print(key, arr22[key])
```

Сформировать 2D массив с распределением Гаусса


```python
arr23, arr24 = np.meshgrid(np.linspace(-1,1,6), np.linspace(-1,1,6))
arr25 = np.hypot(arr23, arr24)
SIGMA, MU = 1.0, 0.0
arr26 = 1/(SIGMA*(2*np.pi)**0.5)*np.exp(-((arr25 - MU) ** 2 / (2.0 * SIGMA ** 2)))
print(arr26)
```
