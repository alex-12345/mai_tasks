```python
import pandas as pd
import numpy as np 
from ipywidgets import IntProgress
from IPython.display import display
```

```python
df = pd.read_csv("fmt.txt", skiprows=1, sep=', ', usecols=[3,5],engine='python',names=['Name','Columns'])
df =df.set_index('Name')
df
```

```python
data = dict()

for index, row in df.iterrows():
    data[index] = []
```

```python
from tqdm import tqdm
from tqdm import tnrange, tqdm_notebook
cnt = 0

with open("log.txt") as f:
    for line in tqdm(f):
        arr = line.rstrip('\n\r').split(', ')
        if arr[0] in data:
            data[arr[0]].append(arr[1:])
        cnt += 1
```

```python
data.keys()
```

```python
for index, row in df.iterrows():
    if(len(data[index]) > 0):
        data[index] = pd.DataFrame(data[index], columns=row.Columns.split(','))
    else:
        del data[index]
```
