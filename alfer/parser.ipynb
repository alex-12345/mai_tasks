```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_rows=200
pd.set_option('display.max_colwidth', None)
_df0 = pd.read_csv("fmt.txt", skiprows=1, sep=', ', usecols=[3,5],engine='python',names=['Name','Columns'])
_df0 = _df0.set_index('Name')
```

```python
data = dict()

for index, row in _df0.iterrows():
    data[index] = []
```

```python
from tqdm import tqdm
from tqdm import tnrange, tqdm_notebook
cnt = 0

with open("log.txt") as f:
    for line in tqdm(f):
        arr = line.rstrip('\n\r,').split(', ')
        if arr[0] in data:
            data[arr[0]].append(arr[1:])
        cnt += 1
```

```python
for index, row in _df0.iterrows():
    if(len(data[index]) > 0):
        data[index] = pd.DataFrame(data[index], columns=row.Columns.split(','))
    else:
        _df0 = _df0.drop(index)
        del data[index]
```

```python
df = data["IMU"][["TimeUS","AccX","AccY","AccZ"]].astype({'TimeUS': 'int32', 'AccX': 'float', 'AccY': 'float', 'AccZ': 'float'}).sort_values(by=['TimeUS'])
df
```

```python
fig = plt.figure(figsize=(18, 9))
ax = fig.add_subplot(111)
ax.scatter(df["TimeUS"].values,df["AccX"].values, s=1, color="red", label='IMU AccX')
ax.scatter(df["TimeUS"].values,df["AccY"].values, s=1, color="green", label='IMU AccY')
ax.scatter(df["TimeUS"].values,df["AccZ"].values, s=1, color="blue", label='IMU AccZ')
ax.legend();
```

```python
acc = (df["AccX"].values**2 + df["AccY"].values**2 + df["AccZ"].values**2)**0.5
fig = plt.figure(figsize=(18, 9))
ax = fig.add_subplot(111)
ax.scatter(df["TimeUS"].values,acc, s=1, color="red", label='IMU Acc')
ax.legend();
```
